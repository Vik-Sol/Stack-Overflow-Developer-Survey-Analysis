```python
#Завантаження бібліотеки pandas
import pandas as pd
```

```python
#Зчитування датасетів для роботи (датасети попередньо завантажені в програму)
survey_results_public = pd.read_csv('survey_results_public.csv', encoding = 'utf-8', low_memory=False)
survey_results_schema = pd.read_csv('survey_results_schema.csv', encoding = 'utf-8', low_memory=False)
```

```python
#Налаштування параметрів для відтворення усіх колонок таблиць
pd.set_options = ('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
#Виводимо назви стовпців для зручності роботи (копіюємо з виведеної таблиці)
for col in survey_results_public.columns:
    print(col)
```

*****************************************************************************************

**************** Завдання 1. Підрахунок загальної кількості респондентів ****************

```python
#Завдання 1. Підрахунок загальної кількості респондентів
total_respondents = survey_results_public['ResponseId'].nunique()
print(f'Загальна кількість респондентів: {total_respondents}')
```

*****************************************************************************************

**************** Завдання 2. Аналіз повноти відповідей респондентів ****************

```python
#Cисок питань з файлу survey_results_schema (поле "qname")
questions_schema = set(survey_results_schema['qname'].unique())

#Cисок питань з файлу survey_results_public (в колонках)
questions_public = set(survey_results_public.columns)

#Пошук спільних питань для обох таблиць
common_questions = questions_schema.intersection(questions_public)

#Актуалізація таблиці survey_results_schema відповідно до спільних питань
survey_results_public_new = survey_results_public[list(common_questions)]

#Підрахунок кількості респондентів, які відповіли на усі питання
num_resp_all_quest = survey_results_public_new.dropna().shape[0]

#Виведення результату
print(f"Кількість респондентів, які відповіли на всі запитання: {num_resp_all_quest}")
```

*****************************************************************************************

**************** Завдання 3. Статистичний аналіз досвіду респондентів ****************

```python
#Видаляємо пропущені значення
work_exp = survey_results_public['WorkExp'].dropna()

#Розраховуємо і виводимо результати
print(f'Статистичний аналіз досвіду респондентів')
print(f'Середнє значення: mean = {round(work_exp.mean(), 1)}')
print(f'Медіана: median = {round(work_exp.median(), 1)}')
print(f'Мода: mode = {round(work_exp.mode()[0], 1)}')
```

*****************************************************************************************

**************** Завдання 4. Аналіз віддаленої роботи ****************

```python
#Знаходимо назву(и) колонки(ок), в яких зустрічається слово "remote"
columns_remote = [col for col in survey_results_public.columns if survey_results_public[col].astype(str).str.contains('remote', case=False, na=False).any()]
columns_remote
```

```python
#Виводимо унікальні значення колонки 'RemoteWork' (обрана релевантна колонка) для коректного використання назви в розрахунках
survey_results_public['RemoteWork'].unique()
```

```python
#Фільтруємо таблицю за респондентами, які працюють тільки віддалено ('Remote')
survey_results_public_remote = survey_results_public[survey_results_public['RemoteWork']=='Remote']

#Розраховуємо і виводимо кількість респондентів, що працюють віддалено
print(f'Кількість респондентів, що працюють віддалено: {survey_results_public_remote.shape[0]}')
```

*****************************************************************************************

**************** Завдання 5. Визначення популярності Python ****************

```python
#Знаходимо назву(и) колонки(ок), в яких зустрічається слово "python" і обираємо релевантну
column_python = [col for col in survey_results_public.columns if survey_results_public[col].astype(str).str.contains('Python', case=False, na=False).any()]
column_python
```

```python
#Знаходимо кількість респондентів, які мають досвід роботи в python (колонка 'LanguageHaveWorkedWith')
python_users = survey_results_public['LanguageHaveWorkedWith'].str.contains('Python', case = False, na = False).sum()

#Знаходимо загальну кількість респондентів
total_users = len(survey_results_public)

#Розраховуємо та виводимо відсоток респондентів, які програмують на Python
python_users_per_cent = round(python_users / total_users * 100, 1)
print (f'Відсоток респондентів, які програмують на Python: {python_users_per_cent} %')
```

**************** Завдання 6. Аналіз шляхів навчання програмуванню ****************

```python
#Знаходимо назву(и) колонки(ок), в яких зустрічається слово "online" і обираємо релевантну
columns_learned_online = [col for col in survey_results_public.columns if survey_results_public[col].astype(str).str.contains('online courses', case=False, na=False).any()]
columns_learned_online
```

*****************************************************************************************

**************** Завдання 7. Географічний аналіз компенсації Python-розробників ****************

```python
#Знаходимо назву(и) стовпця(ів), в якому(их) є частинка слова 'country', щоб обрати релевантний
columns_country = [col for col in survey_results_public.columns if "countr" in col.lower()]
print(columns_country)
```

```python
#Відфільтрувати респондентів, які програмують на Python (релевантна назва колонки була визначена в завданні № 5)
df_python_geogr_comp = survey_results_public[survey_results_public['LanguageHaveWorkedWith'].str.contains('python', case = False, na = False)]

#Залишаємо лише ті колонки, з якими будемо працювати і видаляємо пусті значення для стовпчика ConvertedCompYearly
df_python_geogr_comp = df_python_geogr_comp[['Country', 'ConvertedCompYearly']].dropna(subset=['ConvertedCompYearly'])

#Групуємо дані по країні і розраховуємо середню суму компенсації та медіанну суму компенсації для кожної країни
df_python_geogr_comp_stat = df_python_geogr_comp.groupby('Country')['ConvertedCompYearly'].agg(mean_comp = 'mean', median_comp = 'median').reset_index().round(0)

print(f'Географічний аналіз компенсації Python-розробників')
print(df_python_geogr_comp_stat.to_string())
```

**************** Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів ****************

```python
#Створюємо датасет з потрібними стовпцями відповідно до завдання і видаляємо незаповненні значення компенсації
edlevel_top_spec = survey_results_public[['ResponseId', 'EdLevel', 'ConvertedCompYearly']].dropna(subset=['ConvertedCompYearly'])

#Сортуємо датасет за спаданням за стовпецм ConvertedCompYearly і залишаємо 5 перших рядків (топ-5 найбільш оплачуваних спеціалістів)
edlevel_top5_spec = edlevel_top_spec.sort_values(by = 'ConvertedCompYearly', ascending = False).reset_index(drop=True).head(5)

#Виводимо результат
print(f'Рівні освіти топ-5 найбільш оплачуваних спеціалістів:')
print(edlevel_top5_spec[['EdLevel']])
```
